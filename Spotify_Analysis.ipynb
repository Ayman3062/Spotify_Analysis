### ðŸ“Œ Grouped analysis
We group the data by a categorical column and compute aggregated values (e.g., mean).

```python
average = df.groupby('genre')[['energy','danceability']].mean().reset_index()
```

### ðŸ“Œ Visualization
We generate a chart/plot to better understand trends or comparisons in the data.
Insights

1. Genres with High Energy & Danceability (Top Party Genres):

Reggaeton, Rap, Hip-Hop, Pop, Ska

These genres score high on both energy and danceability, making them popular for clubs, parties, and workouts.

2. Genres with High Danceability but Moderate Energy:

R&B, Soul, Dance, Electronic

They are groovy and smooth, great for dancing but not too aggressive.

3. Genres with High Energy but Lower Danceability:

Rock, Metal-like genres (if included), Opera (surprisingly spiky)

They are powerful and intense, but not necessarily made for dancing.

4. Low Energy & Low Danceability (Calm Genres):

Classical, Opera, Soundtrack, Childrenâ€™s Music

These are more relaxing, soothing, or thematic genres rather than dance tracks.

5. Special Observations:

Anime music sits around moderate energy and moderate danceability.

Jazz has moderate danceability but lower energy, reflecting its smooth, laid-back style.

World Music is balanced but slightly on the lower side for both.

```python

plt.figure(figsize=(12,6))
plt.plot(average['genre'], average['energy'], label='Energy')
plt.plot(average['genre'], average['danceability'], label='Danceability')
plt.grid(True)
plt.xticks(rotation=45)
plt.xlabel("Genre")
plt.ylabel("Average Value")
plt.title("AverageEnergy, and Danceability by Genre")
plt.legend()
plt.show()
```
